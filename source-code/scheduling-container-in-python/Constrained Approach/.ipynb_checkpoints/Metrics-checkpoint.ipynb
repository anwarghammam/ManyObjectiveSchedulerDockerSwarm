```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sat Nov  6 11:40:21 2021

@author: anwar
"""
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Tue May 11 16:25:37 2021

@author: anwar
"""

import os
from instance.Instance import Instance

from Problem.Instance_from_Json import createInstance
from jmetal.algorithm.multiobjective.nsgaiii import UniformReferenceDirectionFactory
from jmetal.operator.crossover import IntegerSBXCrossover
from jmetal.operator.mutation import IntegerPolynomialMutation
from Problem.problem import ReschedulingProblem
from jmetal.util.termination_criterion import StoppingByEvaluations
from jmetal.util.solution import get_non_dominated_solutions, print_function_values_to_file,print_variables_to_file,print_function_values_to_screen,print_variables_to_screen


from jmetal.algorithm.multiobjective.ibea import IBEA
from jmetal.algorithm.multiobjective.nsgaiii import NSGAIII
from jmetal.algorithm.multiobjective.nsgaii import NSGAII
from jmetal.algorithm.multiobjective.random_search import RandomSearch




instance=Instance()
Instance=createInstance(instance)

problem=ReschedulingProblem(instance)


def transform():
    ALGO1= NSGAIII(
                            problem=problem,
                            population_size=200,
                            reference_directions=UniformReferenceDirectionFactory(4,n_partitions=4),
                            mutation=IntegerPolynomialMutation(probability=0.9),
                            crossover=IntegerSBXCrossover(probability=0.05),
                            termination_criterion=StoppingByEvaluations(max_evaluations=2500)
         )
        
    ALGO2=NSGAII(
                        problem=problem,
                        population_size=100,
                        offspring_population_size=100,
                        mutation=IntegerPolynomialMutation(probability=0.9),
                        crossover=IntegerSBXCrossover(probability=0.05),
                        termination_criterion=StoppingByEvaluations(max_evaluations=2500)
                        )

    ALGO3= IBEA(
                        problem=problem,
                        kappa=1.,
                        population_size=200,
                        offspring_population_size=200,
                        mutation=IntegerPolynomialMutation(probability=0.9),
                        crossover=IntegerSBXCrossover(probability=0.05),
                        termination_criterion=StoppingByEvaluations(max_evaluations=2500)
                        )
    ALGO4= algorithm1 = RandomSearch(
                          problem=problem,
                          termination_criterion=StoppingByEvaluations(max_evaluations=2500)
                          )

    os.makedirs(r"/users/anwar/Experiments/NSGAIII")
    os.makedirs(r"/users/anwar/Experiments/NSGAII")
    os.makedirs(r"/users/anwar/Experiments/IBEA")
    os.makedirs(r"/users/anwar/Experiments/random")
    all1=[]
    all2=[]
    all3=[]
    all4=[]
    time1=0
    time2=0
    time3=0
    time4=0

    for i in range(10):
                        #NSGAIII
                        print(i)
                        ALGO1.run()
                        front1 = get_non_dominated_solutions(ALGO1.get_result())
                        
                        for j in front1:
                            all1.append(j)
                       
                        print_function_values_to_file(front1,r"/users/anwar/Experiments/NSGAIII"+"/fun"+str(i))
                        print_variables_to_file(front1, r"/users/anwar/Experiments/NSGAIII/"+"/var"+str(i))
                        time1+=ALGO1.total_computing_time
                        
                        #NSGAII
                       
                        ALGO2.run()
                        front2 = get_non_dominated_solutions(ALGO2.get_result())
                        for j in front2:
                            all2.append(j)
                        print_function_values_to_file(front2,r"/users/anwar/Experiments/NSGAIII"+"/fun"+str(i))
                        print_variables_to_file(front2, r"/users/anwar/Experiments/NSGAIII/"+"/var"+str(i))
                        time2+=ALGO2.total_computing_time
                        
                        
                        #IBEA
                        ALGO3.run()
                        front3 = get_non_dominated_solutions(ALGO3.get_result())
                        for j in front3:
                            all3.append(j)
                        print_function_values_to_file(front3,r"/users/anwar/Experiments/NSGAIII"+"/fun"+str(i))
                        print_variables_to_file(front3, r"/users/anwar/Experiments/NSGAIII/"+"/var"+str(i))
                        time3+=ALGO3.total_computing_time
                        
                        
                        ALGO4.run()
                        front4 = get_non_dominated_solutions(ALGO3.get_result())
                        for j in front4:
                            all4.append(j)
                       
                        print_function_values_to_file(front4,r"/users/anwar/Experiments/random"+"/fun"+str(i))
                        print_variables_to_file(front4, r"/users/anwar/Experiments/random/"+"/var"+str(i))
                        time4+=ALGO4.total_computing_time
                        
                        
    all1= get_non_dominated_solutions(all1)   
    print_function_values_to_file(all1,r"/users/anwar/Experiments/NSGAIII"+"/all1")
    all2= get_non_dominated_solutions(all2,r"/users/anwar/Experiments/NSGAII"+"/all2")
    all3= get_non_dominated_solutions(all3,r"/users/anwar/Experiments/IBEA"+"/all3")
    all4= get_non_dominated_solutions(all2,r"/users/anwar/Experiments/random"+"/all4")     
    print(time1, time2, time3, time4)
```
